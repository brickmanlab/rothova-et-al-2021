# 12 Endoderm explorer subcluster

We take the original endoderm explorer and subcluster it (@alex). 

```python
import scanpy as sc
import matplotlib.pyplot as plt

from pylab import rcParams
```

```python
orig = sc.read("../data/external/nowotschin_et_al_2019/sc_endoderm_all_cells.h5ad")
kat = sc.read("../data/processed/2_Kat.h5ad")
```

```python
kat_subset = orig[kat.obs.barcode]
kat_subset.obs = kat.obs
```

```python
sc.pp.normalize_total(kat_subset, target_sum=10_000)
sc.pp.log1p(kat_subset)
```

```python
rcParams["figure.figsize"] = [20, 5]
sc.pl.dotplot(kat_subset, ['Sfrp1', 'Lefty1', 'Lhx1', 'Cer1', 'Fgf5', 'Dkk1', 'Fzd5', 'Hhex', 'Otx2', 'Hesx1'], groupby='CellType', standard_scale='var')
```
